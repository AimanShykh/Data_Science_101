```python
from sklearn.model_selection import train_test_split
df["target_name"] = df.target.apply(lambda x: digit.target_names[x])
df.head()
```

```python
df0=df[df["target"] == 1]
df1=df[df["target"] == 2]
df2=df[df["target"] == 3]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('pixel_1_1')
plt.ylabel('pixel_7_7')
plt.scatter(df0['pixel_1_1'], df0['pixel_7_7'],color="green",marker='+')
plt.scatter(df1['pixel_1_1'], df1['pixel_7_7'],color="blue",marker='.')
```

```python
plt.xlabel('pixel_2_2')
plt.ylabel('pixel_2_3')
plt.scatter(df0['pixel_2_2'], df0['pixel_2_3'],color="green",marker='+')
plt.scatter(df1['pixel_2_2'], df1['pixel_2_3'],color="blue",marker='.')
```

```python
X = df.drop(["target" , "target_name"] , axis = "columns")
y = df.target
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

**Train Using Support Vector Machine (SVM)**

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train , y_train)
```

```python
model.score(X_test , y_test)
```

```python
model.predict(X_test)
```
