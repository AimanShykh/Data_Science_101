```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
digit = load_digits()
digit
```

```python
digit.target_names
```

```python
digit.target
```

```python
df = pd.DataFrame(digit.data , columns =digit.feature_names)
```

```python
df.head()
```

```python
df["target"] = digit.target
df.head()
```

```python
df[df["target"] == 1]
```

```python
df[df["target"] == 2].head()
```

```python
len(df)
```

```python
from sklearn.model_selection import train_test_split
df["target_name"] = df.target.apply(lambda x: digit.target_names[x])
df.head()
```

```python
df.index[0]
```

```python
X = df.drop(["target" , "target_name"] , axis = "columns")
y = df.target
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train , y_train)
```

```python
model.score(X_test , y_test)
```

```python
model.predict(X_test)
```
