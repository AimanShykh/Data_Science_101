```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data.info()
```

From here, we noticed that Close and Volume are of type object, So this needs to be fixed.

```python
data['Close'].unique()
data['Close'] = data['Close'].str.replace(',','')
data['Close'] = data['Close'].astype(float)
data['Close'].unique()
```

```python
data['Volume'].unique()
data['Volume'] = data['Volume'].str.replace(',','')
data['Volume'] = data['Volume'].astype(int)
data['Volume'].unique()
```

```python
data.describe()
```

```python
sort1=data.sort_values(by="Open")
sort2=data.sort_values(by="High")
sort3=data.sort_values(by="Low")
sort4=data.sort_values(by="Close")
sort5=data.sort_values(by="Volume")
```

# Data Visualization

```python
sns.pairplot(data)
```

```python
plt.subplots(figsize=(12, 6))
plt.plot(data.index, data["Open"])

# Customize the plot
plt.xlabel('Date')
plt.ylabel("Open")
plt.title('Time Series Plot')
plt.show()
```

```python
data.Open.nlargest(20).plot(kind='bar',figsize=(10,5),color='blue', position=0, width=0.4)
data.Close.nlargest(20).plot(kind='bar',figsize=(10,5),color='red', position=1, width=0.4)
plt.title("Google Highest Open and Close Prices")
plt.legend()
plt.ylabel('Value')
plt.show()
```

```python
plt.subplots(figsize=(10,6))

plt.scatter(data['Open'], data['Volume'])
plt.title('Open Price and Volume')
plt.xlabel('Open Price')
plt.ylabel('Volume')
plt.show()
```
